**Question 1**

A permutation perm of n + 1 integers of all the integers in the range [0, n] can be represented as a string s of length n where:

- s[i] == 'I' if perm[i] < perm[i + 1], and
- s[i] == 'D' if perm[i] > perm[i + 1].

Given a string s, reconstruct the permutation perm and return it. If there are multiple valid permutations perm, return **any of them**.

**Example 1:**

**Input:** s = "IDID"

**Output:**

[0,4,1,3,2]

```python
import java.util.*;
public class DI_str_match {
    public static int[] diStringMatch(String s) {
        int n = s.length();
        int N = s.length();
        int k=0;
        int ans[] = new int[N+1];
       for(int i=0;i<N;i++){
           if(s.charAt(i)=='I'){
               ans[i]=k++;
           }
           else{
               ans[i]=n--;
           }
       } 
       ans[N]=n;
       return ans;
    }
    public static void main(String args[]){
        Scanner sc = new Scanner(System.in);
        String s = sc.nextLine();
        int ans[] = diStringMatch(s);
        for(int i=0;i<ans.length;i++){
            System.out.print(ans[i]+",");
        }
    }
}
```

**Question 2**

You are given an m x n integer matrix matrix with the following two properties:

- Each row is sorted in non-decreasing order.
- The first integer of each row is greater than the last integer of the previous row.

Given an integer target, return true *if* target *is in* matrix *or* false *otherwise*.

You must write a solution in O(log(m * n)) time complexity.

**Example 1:**
**Input:** matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 3

**Output:** true

```python
import java.util.*;
public class Search_2D_matrix {
    public static boolean searchMatrix(int[][] matrix, int target) {
        int k=matrix.length;
        int l=matrix[0].length;
        int st=0 , end=k*l-1;
        while(st<=end){
            int mididx,midele,row,col;
            mididx = st+(end-st)/2;
            row = mididx/l;
            col = mididx%l;
            midele = matrix[row][col];
            if(midele==target){
                return true;
            }
            else if(midele<target){
                st=mididx+1;
            }
            else{
                end=mididx-1;
            }
        }
        return false;
    }
    public static void main(String args[]){
        Scanner sc = new Scanner(System.in);
        int m=sc.nextInt();
        int n= sc.nextInt();
        int arr[][]= new int[m][n];
        for(int i=0;i<m;i++){
            for(int j=0;j<n;j++){
                arr[i][j]=sc.nextInt();
            }
        }
        int target=sc.nextInt();
        System.out.println(searchMatrix(arr,target));
    }
}
```

**Question 3**

Given an array of integers arr, return *true if and only if it is a valid mountain array*.

Recall that arr is a mountain array if and only if:

- arr.length >= 3
- There exists some i with 0 < i < arr.length - 1 such that:
    - arr[0] < arr[1] < ... < arr[i - 1] < arr[i]
    - arr[i] > arr[i + 1] > ... > arr[arr.length - 1]
**Example 1:**

**Input:** arr = [2,1]

**Output:**

false

```python
import java.util.*;
public class Valid_Mountain {
    public static boolean validMountainArray(int[] arr) {
        int mx=Integer.MIN_VALUE;
        int d=-1,k=0,l=0;
        for(int i=0;i<arr.length;i++){
            if(arr[i]>mx){
                mx=arr[i];
                d=i;
            }
        }
        for(int i=1;i<d+1;i++){
            if(arr[i-1]>=arr[i]){
                return false;
            }
            k=1;
        }
        for(int i=d;i<arr.length-1;i++){
            if(arr[i]<=arr[i+1]){
                return false;
            }
            l=1;
        }
        if(k==0 || l==0){
            return false;
        }
        return true;
    }
    public static void main(String args[]){
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[]=new int[n];
        for(int i=0;i<n;i++){
            arr[i] = sc.nextInt();
        }
        System.out.println(validMountainArray(arr));
    }
}
```

**Question 4**

Given a binary array nums, return *the maximum length of a contiguous subarray with an equal number of* 0 *and* 1.

**Example 1:**

**Input:** nums = [0,1]

**Output:** 2

**Explanation:**

[0, 1] is the longest contiguous subarray with an equal number of 0 and 1.

</aside>

```python
import java.util.*;
public class Contiguous_Array01 {
    public static int findMaxLength(int[] nums) {
        int arr[] = new int[2*nums.length+1];
        Arrays.fill(arr,-2);
        arr[nums.length]=-1;
        int count=0,mx=0;
        for(int i=0;i<nums.length;i++){
            count=count+(nums[i]==0?-1:1);
            if(arr[count+nums.length]>=-1){
                mx=Math.max(mx,i-arr[count+nums.length]);
            }
            else{
                arr[count+nums.length] = i;
            }
        }
        return mx;
    }
    public static void main(String args[]){
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[]=new int[n];
        for(int i=0;i<n;i++){
            arr[i] = sc.nextInt();
        }
        System.out.println(findMaxLength(arr));
    }
}
```

**Question 5**

The **product sum** of two equal-length arrays a and b is equal to the sum of a[i] * b[i] for all 0 <= i < a.length (**0-indexed**).

- For example, if a = [1,2,3,4] and b = [5,2,3,1], the **product sum** would be 1*5 + 2*2 + 3*3 + 4*1 = 22.

Given two arrays nums1 and nums2 of length n, return *the **minimum product sum** if you are allowed to **rearrange** the **order** of the elements in* nums1.

**Example 1:**

**Input:** nums1 = [5,3,4,2], nums2 = [4,2,2,5]

**Output:** 40

**Explanation:**

We can rearrange nums1 to become [3,5,4,2]. The product sum of [3,5,4,2] and [4,2,2,5] is 3*4 + 5*2 + 4*2 + 2*5 = 40.

```python
import java.util.*;
public class Min_prodsum_2_arr {
    public static int prod(int arr1[],int arr2[]){
        int n=arr1.length;
        Arrays.sort(arr2);
        Arrays.sort(arr1);
        int sum=0;
        for(int i=0;i<n;i++){
            sum=sum+arr1[i]*arr2[n-i-1];
        }
        return sum;
    }
    public static void main(String args[]){
    Scanner sc = new Scanner(System.in);
    int n=sc.nextInt();
    int arr1[] = new int[n];
    for(int i=0;i<n;i++){
        arr1[i] = sc.nextInt();
    }
    int arr2[] = new int[n];
    for(int i=0;i<n;i++){
        arr2[i] = sc.nextInt();
    }
    int ans=prod(arr1,arr2);
    System.out.println(ans);
}
}
```

**Question 6**

An integer array original is transformed into a **doubled** array changed by appending **twice the value** of every element in original, and then randomly **shuffling** the resulting array.

Given an array changed, return original *if* changed *is a **doubled** array. If* changed *is not a **doubled** array, return an empty array. The elements in* original *may be returned in **any** order*.

**Example 1:**

**Input:** changed = [1,3,4,2,6,8]

**Output:** [1,3,4]

**Explanation:** One possible original array could be [1,3,4]:

- Twice the value of 1 is 1 * 2 = 2.
- Twice the value of 3 is 3 * 2 = 6.
- Twice the value of 4 is 4 * 2 = 8.

Other original arrays could be [4,3,1] or [3,1,4].

```python
import java.util.*;
public class Original_arr_frm_Double_arr {
    public static int[] findOriginalArray(int changed[]) {
        if(changed.length%2==1){
            return new int[0];
        }
        int ans[] = new int[changed.length/2];
        int count[] = new int[100001];
        for(int i:changed){
            count[i]++;
        }
        int idx=0;
        for(int i=0;i<100001;i++){
            while(count[i]>0 && (i*2)<100001 && count[i*2]>0){
                count[i]--;
                count[i*2]--;
                ans[idx]=i;
                idx++;
            }
        }
        for(int i=0;i<100001;i++){
            if(count[i]>0){
                return new int[0];
            }
        }
        return ans;
    }
    public static void main(String args[]){
        Scanner sc = new Scanner(System.in);
        int n=sc.nextInt();
        int arr[] = new int[n];
        for(int i=0;i<n;i++){
            arr[i] = sc.nextInt();
        }
        int ans[] = findOriginalArray(arr);
        for(int i=0;i<ans.length;i++){
            System.out.print(ans[i]+",");
        }
    }
}
```

**Question 7**

Given a positive integer n, generate an n x n matrix filled with elements from 1 to n2 in spiral order.

**Example 1:**

**Input:** n = 3

**Output:** [[1,2,3],[8,9,4],[7,6,5]]

```python
import java.util.*;
public class Spiral_Matrix2 {
      public static int[][] generateMatrix(int n) {
        int arr[][]=new int[n][n];
        int cnt=1;
        for(int i=0;i<(n+1)/2;i++){
            for(int j=i;j<n-i;j++){
                arr[i][j]=cnt++;
            }
            for(int j=i+1;j<n-i;j++){
                arr[j][n-i-1]=cnt++;
            }
            for(int j=i+1;j<n-i;j++){
                arr[n-i-1][n-j-1]=cnt++;
            }
            for(int j=i+1;j<n-i-1;j++){
                arr[n-j-1][i]=cnt++;
            }
        }
        return arr;
    }
    public static void main(String args[]){
        Scanner sc = new Scanner(System.in);
        int n=sc.nextInt();
        int arr[][]=generateMatrix(n);
        for(int i=0;i<n;i++){
            for(int j=0;j<n;j++){
                System.out.print(arr[i][j]+" ");
            }
            System.out.println();
        }
    }
}
```

**Question 8**

Given two [sparse matrices](https://en.wikipedia.org/wiki/Sparse_matrix) mat1 of size m x k and mat2 of size k x n, return the result of mat1 x mat2. You may assume that multiplication is always possible.

**Example 1:**

</aside>
**Input:** mat1 = [[1,0,0],[-1,0,3]], mat2 = [[7,0,0],[0,0,0],[0,0,1]]

**Output:**

[[7,0,0],[-7,0,3]]

```python
import java.util.*;
public class SparseMatrix {
    public static int[][] sparse(int arr1[][],int arr2[][],int m,int l,int n){
        int sm=0;
        int ans[][]= new int[m][n];
        for(int k=0;k<m;k++){
            for(int i=0;i<n;i++){
                sm=0;
                for(int j=0;j<l;j++){
                    sm=sm+arr1[k][j]*arr2[j][i];
                }
                ans[k][i]=sm;
            }
        }
        return ans;
    }
    public static void main(String args[]){
        Scanner sc = new Scanner(System.in);
        int m=sc.nextInt();
        int k=sc.nextInt();
        int n= sc.nextInt();
        int arr1[][] = new int[m][k];
        int arr2[][] = new int[k][n];
        for(int i=0;i<m;i++){
            for(int j=0;j<k;j++){
                arr1[i][j]=sc.nextInt();
            }
        }
        for(int i=0;i<k;i++){
            for(int j=0;j<n;j++){
                arr2[i][j]=sc.nextInt();
            }
        }
        int ans[][] = sparse(arr1,arr2,m,k,n);
        for(int i=0;i<m;i++){
            for(int j=0;j<n;j++){
                System.out.print(ans[i][j]+" ");
            }
            System.out.println();
        }
    }
}
```
